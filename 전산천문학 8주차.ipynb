```python
#midpoint RK method

N=1000
h=10/N
y=[3]
x=[0]
y2=[3]
y3=[0]
def func(x,y):
    return -2*x+y

for i in range(0,N):
    y.append(y[i]+h*func(x[i]+h/2,y[i]+(h/2)*(func(x[i],y[i]))))
    y2.append(2+2*x[i]+np.exp(x[i])) 
    y3.append(np.abs((y2[i]/y[i])-1))
    x.append(x[i]+h)
    
plt.yscale('log')    
plt.plot(x,y,label='RK')
plt.plot(x,y2,label='original')
plt.legend(loc=2)
```

```python
#RK4

N=10000
h=10/N
Y=[y]
X=[x]
x,y=0,2

def func(x,y):
    return -y+np.cos(x)-np.sin(x)

def rk4(x,y,h):
    v1=func(x,y)
    v2=func(x+0.5*h,y+0.5*h*v1)
    v3=func(x+0.5*h,y+0.5*h*v2)
    v4=func(x+h,y+h*v3)
    return y+(h/6.)*(v1+2*v2+2*v3+v4)

for i in range (0,N):
    y=rk4(x,y,h)
    x=x+h
    Y=np.append(Y,y)
    X=np.append(X,x) #np.append 사용법에 주의
    
plt.plot(X,Y,'c.')
```
