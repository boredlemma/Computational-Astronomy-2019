```python
#1
def v(d,N):
    
    num=0
    for j in np.arange(0,N):
        r=0
        for i in range(0,d):
            x=np.random.rand()
            r+=x**2
        if np.sqrt(r)<=1:
            num+=1
    return (num/N)*2**d
```

```python
X=[]
d=3
N=pow(10,7)
for i in range(0,100):
    X.append(v(d,N))
print(np.mean(X),np.std(X))
true=(4/3)*np.pi
print('the ratio with regard to the real value obtained from the given formula=',np.mean(X)/(true))
```

```python
#2(a) taumax=0.01

taumax=pow(10,-2)
n=pow(10,5)
refl,trans=0,0


for i in range(0,n):
    x=0
    y=0
    z=0
    
    while (0<=z<1):
        
        xi4=np.random.rand()
        xi5=np.random.rand()

        mu=2*xi5-1 #cos(theta)
        phi=2*np.pi*xi4
        mu2=np.sqrt(1-mu**2) #sin(theta)

    
        xi3=np.random.rand()
        tau=-np.log(xi3)
        P=1-np.exp(-tau)
        L=tau/taumax
        
        x+=L*mu2*np.cos(phi)
        y+=L*mu2*np.sin(phi)
        z+=L*mu
      
    
        if z<0:
            refl+=1
            break
        elif z>=1:
            trans+=1
            break
        else:
            continue
        

print(refl/n,trans/n)
```

```python
#3
def dudt(u,x):
    X=(2/x)-1/(x**2)
    Y=u-(1/u)
    return X/Y
```

```python
#3 (b) +2

def func2(x,u):
    if  u==1 and x== 1/2:
        return 2
    
    else:
        X=(2/x)-1/(x**2)
        Y=u-(1/u)
        return X/Y


def rk4(x,u,h):
    v1=func2(x,u)
    v2=func2(x+0.5*h,u+0.5*h*v1)
    v3=func2(x+0.5*h,u+0.5*h*v2)
    v4=func2(x+h,u+h*v3)
    return u+(h/6.)*(v1+2*v2+2*v3+v4)



x,u=0.5,1
N=10000
h=(-0.5+5)/N
U=[u]
X=[x]

for i in range (0,N-1):
    u=rk4(x,u,h)
    x=x+h
    U=np.append(U,u)
    X=np.append(X,x) 
plt.plot(U,X,'#349bf1')

x,u=0.5,1
N=1000
h=(0.1-0.5)/N
U=[u+2*h]
X=[x+h]

for i in range (0,N):
    u=rk4(x,u,h)
    x=x+h
    U=np.append(U,u)
    X=np.append(X,x) 
    
plt.plot(U,X,'#349bf1', label='+2')

#3 (b) -2

def func2(x,u):
    if  u==1 and x== 1/2:
        return -2
    
    else:
        X=(2/x)-1/(x**2)
        Y=u-(1/u)
        return X/Y


def rk4(x,u,h):
    v1=func2(x,u)
    v2=func2(x+0.5*h,u+0.5*h*v1)
    v3=func2(x+0.5*h,u+0.5*h*v2)
    v4=func2(x+h,u+h*v3)
    return u+(h/6.)*(v1+2*v2+2*v3+v4)



x,u=0.5,1
N=10000
h=(-0.5+5)/N
U=[u]
X=[x]

for i in range (0,N-1):
    u=rk4(x,u,h)
    x=x+h
    U=np.append(U,u)
    X=np.append(X,x) 
plt.plot(U,X,'#61E4F9')

x,u=0.5,1
N=1000
h=(0.1-0.5)/N
U=[u+2*h]
X=[x+h]

for i in range (0,N):
    u=rk4(x,u,h)
    x=x+h
    U=np.append(U,u)
    X=np.append(X,x) 
    
plt.plot(U,X,'#61E4F9',label='-2')

plt.legend(loc=1)


```

```python
#4(b)

def deriv(y,x):
    n=0
    dydt0=y[1]
    dydt1=-(2*x*y[1]+(x**2)*(y[0]**n))/(x**2)
    return [dydt0,dydt1]

yini=np.array([1,0])
xini=pow(10,-10)
xmax=5
n=100
x=np.linspace(xini,xmax,n)
y=odeint(deriv,yini,x)
z=1-(1/6)*(x**2)

fig=plt.figure(1,figsize=(9,6))
plt.plot(x,y[:,0],label='computational',c='#0B31AF')
plt.plot(x,z,'-.',label='analytic',c='#dceffd')
plt.legend(loc=1)


fig=plt.figure(2,figsize=(6,4))
w=np.abs((z-y[:,0])/y[:,0])
plt.plot(x,w,label='error',c='#73c2fb')
plt.legend(loc=1)

```
