## 3 

```python
def prime(n):
    if n==2:
        return True
    else:
        for m in range(2,n):
            if n%m==0:
                return False
            else:
                continue
        return True
           
nmax=int(input('Enter an integer greater than 1: '))
pnum=0

for i in range (2,nmax+1):
    if prime(i)==True:
        pnum+=1
    else:
        continue
print('There are ',pnum, 'prime numbers between 2 and ',nmax)
```

#### The number of prime numbers from 2 to 100 is 25.

#### The number of prime numbers from 2 to 1000 is 168.

#### The number of prime numbers from 2 to 10000 is 1229.

#### The number of prime numbers from 2 to 100000 is 9592.
